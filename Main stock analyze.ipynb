```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_datareader import data, wb
from datetime import datetime
```

```python
amazon = pd.read_csv("/Users/raghavgupta/Downloads/amzn.csv")
amazon.head()
google = pd.read_csv("/Users/raghavgupta/Downloads/googl.csv")
google.head()
microsoft = pd.read_csv("/Users/raghavgupta/Downloads/msft.csv")
microsoft.head()
facebook = pd.read_csv("/Users/raghavgupta/Downloads/fb.csv")
facebook.head()
apple = pd.read_csv("/Users/raghavgupta/Downloads/aapl.csv")
apple.head()
```

```python
stocks = pd.DataFrame({"AAPL": apple["Close"],
                      "MSFT": microsoft["Close"],
                      "FB": facebook["Close"],
                      "AMZN": amazon["Close"],
                       "GOOGL": google["Close"]})

stocks.plot(grid = True)
```

```python
stocks.plot(secondary_y = ["AMZN"], grid = True)
```

```python
stock_return = stocks.apply(lambda x: x / x[0])
stock_return.plot(grid = True).axhline(y = 1, color = "black", lw = 2)
```

```python
stock_change = stocks.apply(lambda x: np.log(x) - np.log(x.shift(1))) # shift moves dates back by 1.
stock_change.head()
stock_change.plot(grid = True).axhline(y = 0, color = "black", lw = 2)
```

```python
apple['Daily Return'] = apple['Close'].pct_change()
amazon['Daily Return'] = amazon['Close'].pct_change()
facebook['Daily Return'] = facebook['Close'].pct_change()
google['Daily Return'] = google['Close'].pct_change()
microsoft['Daily Return'] = microsoft['Close'].pct_change()
```

```python
sns.distplot(apple['Daily Return'].dropna(), bins = 100,
            color = 'blue')
```
